This script generates png's from the txt scripts generated by the simulation.  
  
We start by loading the data from the isothermal gauss peak simulation.

```python
import numpy as np
import matplotlib.pyplot as plt

def load(filename):
    a = np.loadtxt('q1' + filename, delimiter=',')
    b = np.loadtxt('q2' + filename, delimiter=',')
    c = np.loadtxt('q3' + filename, delimiter=',')
    return a, b, c

q1, q2, q3 = load('gauss0.000000.txt')
q1
```

Now we will generate a plot for each line(save time step) against the Position x

```python
x = np.linspace(0.0, 100.0, 500)


for i in range(len(q1)):
    plt.plot(x, q1[i])
    plt.title('Isothermal inviscid 1D sim density')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$\rho$')
    plt.savefig('Plots/q1gauss/' +str(i).zfill(3) + '.png')
    plt.close()
    plt.plot(x, q2[i])
    plt.title('Isothermal inviscid 1D sim velocity')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$u$')
    plt.savefig('Plots/q2gauss/' +str(i).zfill(3) + '.png')
    plt.close()
```

Now we generate the plots for exercise 1b), the non-isothermal non-viscous case:

```python
q1, q2, q3 = load('jump0.000000.txt')

for i in range(len(q1)):
    plt.plot(x, q1[i])
    plt.title('Non-isothermal inviscid 1D sim density')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$\rho$')
    plt.savefig('Plots/q1/' +str(i).zfill(3) + '.png')
    plt.close()
    plt.plot(x, q2[i])
    plt.title('Non-isothermal inviscid 1D sim velocity')
    plt.xlabel(r'$x$')
    plt.ylabel(r'$u$')
    plt.savefig('Plots/q2/' +str(i).zfill(3) + '.png')
    plt.close()
    plt.plot(x, q3[i])
    plt.title('Non-isothermal inviscid 1D sim energy')
    plt.xlabel(r'$x$')
    plt.ylabel(r'internal energy $\epsilon$')
    plt.savefig('Plots/q3/' +str(i).zfill(3) + '.png')
    plt.close()
```
